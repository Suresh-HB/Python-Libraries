@Title : Write a program to draw box plot of life expectancy by continent for a data set given in a
url

```python
import myloggingfile as mlf
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

logger = mlf.logger_init("seaBoxplot_P6.ipynb")

def box_plot_life_expectancy(url):
    """
    Description: Function to draw a box plot of life expectancy by continent for a given dataset.
    Parameters:
        url : URL to read the data from
    Returns:
        None
    """
    
    logger.info("inside box_plot_life_expectancy method")
    plt.figure(figsize=(12, 8))
    
    data = pd.read_csv(url)
    logger.debug(f"data \n{data.head()}")
    
    sns.boxplot(x='continent', y='lifeExp', data=data, palette='coolwarm')
    plt.title('Box Plot of Life Expectancy by Continent')
    plt.xlabel('Continent')
    plt.ylabel('Life Expectancy')
    plt.xticks(rotation=45)  
    plt.show()
    logger.info("plotting and labeling theboxplot graph")

def main():
    logger.info("inside main method")
    
    url = "https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv"
    logger.info("calling box_plot_life_expectancy method")
    box_plot_life_expectancy(url)

if __name__ == '__main__':
    main()
```

@Title : Write a program to draw a box plot of day by tips for a dataset given in a url

```python
import myloggingfile as mlf
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

logger = mlf.logger_init("seaBoxplot_P6.ipynb")

def box_plot_tips_by_day(url):

    """
    Description: Function to draw a box plot of tips by day for a given dataset.
    Parameters:
        url : url to read the data from file 
    Returns:
        None
    """

    logger.info("inside box_plot_tips_by_day method")
    plt.figure(figsize=(12, 8))
  
    data = pd.read_csv(url)
    logger.debug(f"data \n{data.head()}")
    
    sns.boxplot(x='day', y='tip', data=data, palette='pastel')
    plt.title('Box Plot of Tips by Day')
    plt.xlabel('Day')
    plt.ylabel('Tip')
    plt.xticks(rotation=45)  
    plt.show()
    logger.info("pltting and labeling the graph")

def main():
    logger.info("inside main method")
    
    url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
    logger.info("calling box_plot_tips_by_day method")
    box_plot_tips_by_day(url)

if __name__ == '__main__':
    main()
```
