@Title : program to draw a point plot for sex against survived for a dataset given in url 

```python
import myloggingfile as mlf
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

logger = mlf.logger_init("pointplot_P2.ipynb")


def  bar_plot_against_survived(url):

    """
    Description: Function to draw bar plot of sex against survived for a dataset given in the url.
    Parameters:
        url : Function taking url as a parameter into the function to read data and draw the graph        
    Returns:
        None:
    """
  
    logger.info("inside bar_plot_against_survived method")
    
    data = pd.read_csv(url)
    plt.figure(figsize=(8, 6))
    sns.pointplot(x='sex', y='survived', data=data,errorbar=None, markers='d',color='green')

    plt.title('Point Plot of Sex Against Survival')
    plt.xlabel('Sex')
    plt.ylabel('Survived')
    plt.xticks(rotation=45)
    plt.ylim(-0.1, 1.1)  
    plt.show()

def main():

    logger.info("inside main method")
    url = "https://github.com/mwaskom/seaborn-data/raw/master/titanic.csv"
    logger.info("calling bar_plot_against_survived method")
    bar_plot_against_survived(url)
    
if __name__=='__main__':
    main()
```
