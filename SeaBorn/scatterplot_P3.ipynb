@Title : Write a program to draw a scatter plot of “day” against “total bill” for a dataset given in a
url 

```python
import myloggingfile as mlf
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
logger = mlf.logger_init("scatterplot_P3.ipynb")


def total_bill(url):

    """
    Description: Function to draw a scatter plot of “day” against “total bill” for a dataset 
    Parameters:
        url : Function taking url as a parameter into the function to read data and draw the scotter plot      
    Returns:
        None:
    """
  
    logger.info("inside total_bill method")
    plt.figure(figsize=(10, 6))
    data = pd.read_csv(url)
    logger.debug(data)
    sns.scatterplot(x='day', y='total_bill', data=data, hue='day', palette='viridis', s=100)


    plt.title('Scatter Plot of Day Against Total Bill')
    plt.xlabel('Day')
    plt.ylabel('Total Bill')
    plt.xticks(rotation=45)
    plt.show()


def main():

    logger.info("inside main method")
    url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
    logger.info("calling total_bill method")
    total_bill(url)
    
if __name__=='__main__':
    main()
```
