```python
import matplotlib.pyplot as plt
import pandas as pd
import myloggingfile as mlf

logger = mlf.logger_init("piechartwithlabel_P2.ipynb")

def piechart_with_title(Programming_languages,Popularity):

    """
    Description: Function to create a pie chart of the popularity of programming Languages with title.  
    Parameters:
       Programming_languages, Popularity : Function taking the data as a parameter to draw the pie chart of the popularity of programming Languages. 
    Returns:
        None:
    """

    logger.info("inside piechart_with_title method")
    plt.title("popularity of programming Languages")
    plt.xlabel('Popularity')
    plt.ylabel('Programming_languages')
    plt.pie(Popularity,labels=Programming_languages,startangle=140,shadow=True,autopct='%1.1f%%')
    plt.show()
    logger.info("plotted the piechart with title")

def main():
     
    logger.info("inside main method")
    Programming_languages= ['Java','Python','PHP','JavaScript','C#','C++ ']
    Popularity= [22.2, 17.6, 8.8, 8, 7.7, 6.7] 
    logger.info("calling piechart_with_title method")
    piechart_with_title(Programming_languages,Popularity)
    

if __name__=='__main__':
    main()
```
