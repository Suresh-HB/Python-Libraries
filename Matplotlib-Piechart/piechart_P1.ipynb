```python
import matplotlib.pyplot as plt
import pandas as pd
import myloggingfile as mlf
import numpy as np
logger = mlf.logger_init("piechart_P1.ipynb")


def pie_chart():

    """
    Description: Function to create a pie chart of the popularity of programming Languages.  
    Parameters:
       data : Function taking the data as a parameter to draw the pie chart of the popularity of programming Languages. 
    Returns:
        None:
    """
    
    logger.info("inside pie_chart method")
    languages = 'Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'
    popuratity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
    colors = ["red", "skyblue", "yellow", "green", "blue", "black"]
    explode = (0.1, 0, 0, 0,0,0)  

    k=plt.pie(popuratity, explode=explode, labels=languages, colors=colors,  autopct='%1.1f%%',shadow=True,startangle=140)
    logger.debug(f" plotting the pie chart\n {k}")
    plt.axis('equal')
    plt.show()
    logger.info("plotted the piechart")

def main():
    
    logger.info("inside main method")
    logger.info("calling pie_chart method")
    pie_chart()


if __name__=='__main__':
    main()
```
