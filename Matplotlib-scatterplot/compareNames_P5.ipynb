```python
import matplotlib.pyplot as plt
import numpy as np
import myloggingfile as mlf

logger = mlf.logger_init("compareNames_P5.ipynb")

def comapare_weight_heights():

    logger.info("inside compare_weight_heights method")
    groups=['Group1','group2','group3']
    
    x=[np.random.rand(1,5) for _ in range(3)]
    y=[np.random.rand(1,5) for _ in range(3)]
    logger.info("randomly taking values for each group")
    z={}
    for i in range (3):
        z[i]=[x[i],y[i]]  
        

    c=["#1f77b4", "#ff7f0e", "#2ca02c"]   
    s=['s','o','^'] 
    logger.debug(f"colors and markers {c} and {s}")

    for key ,val in z.items():
        label='Group'+str(key)
        plt.scatter(val[0], val[1], marker=s[key], label=label , color=c[key])
        
    plt.xlabel('Weight (kg)')
    plt.ylabel('Height (cm)')
    plt.title('Scatter Plot of Weights and Heights for Three Different Groups')

    plt.legend()
    plt.grid(True)
    plt.show()
    logger.info("title and labeling thex&y axis and ploted the graph")


def main():

    logger.info("inside main method")
    logger.info("calling different_weight method")
    comapare_weight_heights()

if __name__=='__main__':
    main()
```
