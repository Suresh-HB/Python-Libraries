```python
import matplotlib.pyplot as plt
import numpy as np
import myloggingfile as mlf

logger = mlf.logger_init("scatterplot_P1.ipynb")

def draw_scatter_graph():

    """
    Description: Function to draw the scatter graph taking a random distribution in X and Y and plotted against each other. 
    Parameters:
    Returns:
        None:
    """

    logger.info("inside draw_scatter_graph")
    num_points = 100  
    x = np.random.rand(num_points) * 100 
    logger.debug(x)

    y = np.random.rand(num_points) * 100  
    logger.debug(y)

    plt.figure(figsize=(8, 6))
    plt.scatter(x, y, color='red', marker='o', alpha=0.7)
    logger.debug(f"plotting")


    plt.title('Scatter Plot of Random Data')
    plt.xlabel('X axis')
    plt.ylabel('Y axis')
    plt.grid(True)
    plt.show()
    logger.info("title and labeling the x&y axis and displaying the graph")


def main():
    
    logger.info("inside main method")
    logger.info("calling draw_scatter_graph method")
    draw_scatter_graph()

if __name__=='__main__':
    main()
```
