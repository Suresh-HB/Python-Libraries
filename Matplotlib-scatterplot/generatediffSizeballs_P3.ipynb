```python
import math
import random
import matplotlib.pyplot as plt
import myloggingfile as mlf
logger = mlf.logger_init("generatediffSizeballs_P3.ipynb")

def draw_grpg_balls():

    """
    Description: Function to draw a scatter plot using random distributions to generate balls of
                different sizes.
    Parameters:
    Returns:
        None:
    """

    logger.info("inside draw_grpg_balls method")
    no_of_balls = 25
    logger.debug(f"no of balls taking\n{no_of_balls}")
    x = [random.triangular() for i in range(no_of_balls)]
    y = [random.gauss(0.5, 0.25) for i in range(no_of_balls)]
    colors = [random.randint(1, 4) for i in range(no_of_balls)]
    areas = [math.pi * random.randint(5, 15)**2 for i in range(no_of_balls)]

    plt.figure()
    plt.scatter(x, y, s=areas, c=colors, alpha=0.85)
    plt.axis([0.0, 1.0, 0.0, 1.0])
    plt.xlabel("X")
    plt.ylabel("Y")
    logger.info("plotting and labeling the x & y axis ")
    plt.show()

def main():

    logger.info("inside main method")
    logger.info("calling draw_grpg_balls method")
    draw_grpg_balls()


if __name__=='__main__':
    main()
```
