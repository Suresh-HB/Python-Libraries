```python
import matplotlib.pyplot as plt
import numpy as np
import myloggingfile as mlf
logger = mlf.logger_init("drawEmptyCircles_P2.ipynb")

def draw_empty_circle():

    """
    Description: Function to draw a scatter plot with empty circles taking a random distribution in X and Y 
                and plotted against each other.
    Parameters:
    Returns:
        None:
    """

    logger.info("inside draw_empty_circle method")
    x = np.random.randn(50) 
    y = np.random.randn(50)
    logger.debug(f"taking rando values for X and Y axis {x} & {y}")

    k = plt.scatter(x, y, s=70, facecolors='none', edgecolors='g')
    logger.debug(f"plotting the scatter graph :{k}")
    
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.show()
    logger.info("labeling the x&y axis and displaying the graph")


def main():

    logger.info('inside main method')
    logger.info("calling draw_empty_circle method")
    draw_empty_circle()


if __name__=='__main__':
    main()
```
