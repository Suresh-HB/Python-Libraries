```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import myloggingfile as mlf
logger = mlf.logger_init("lableontoperroebars_P13.ipynb")

def error_bars(mean_velocity,std_deviation,labels):

    """
    Description: Function to create bar plots with error bars on the same figure.
    Parameters:
       mean_velocity, std_deviation, labels : Function taking parameters for plotting bars
    Returns:
        None:
    """

    logger.info("inside error_bars method")
    mean_velocity_int = [int(round(value)) for value in mean_velocity]
    num_bars = len(mean_velocity)
    positions = np.arange(num_bars)
    bar_width = 0.5
    plt.figure(figsize=(10, 6))
    
    k = bars = plt.bar(positions, mean_velocity, yerr=std_deviation, width=bar_width, color='skyblue', edgecolor='black', capsize=5, label='Mean Velocity')
    logger.debug(f"plotting the bars \n{k}")

    plt.title('Bar Plot with Error Bars and Text Labels')
    plt.xlabel('Measurements')
    plt.ylabel('Mean Velocity')
    plt.xticks(positions, labels)
    logger.info("title and labeling x&y axis,xticks")

    for bar, value in zip(bars, mean_velocity_int):
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, height, str(value), ha='center', va='bottom', fontsize=10, color='black')

    plt.legend()
    plt.show()


def main():
    
    logger.info("inside main method")
    mean_velocity = [0.2474, 0.1235, 0.1737, 0.1824]
    std_deviation = [0.3314, 0.2278, 0.2836, 0.2645]
    labels = ['Measure 1', 'Measure 2', 'Measure 3', 'Measure 4']
    logger.info("calling error_bars method")
    error_bars(mean_velocity,std_deviation,labels)


if __name__ == "__main__":
    main()
```
