1. Write a Python programming to display a bar chart of the popularity of programming
Languages.
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

```python
import matplotlib.pyplot as plt
import pandas as pd
import myloggingfile as mlf
logger = mlf.logger_init("popularityProgmLang_P1.ipynb")


def draw_bar_chart(data):

    """
    Description: Function to display a bar chart of the popularity of programming Languages. 
    Parameters:
       data : Function taking the data as a parameter to draw the barchart after make the data into data Frame
    Returns:
        None:
    """

    logger.info("inside draw_bar_chart")
    df = pd.DataFrame(data)
    logger.info(f" Data Frame\n{df}")

    plt.bar(df['programming'],df['popularity'],label='popularity',color='green',width=0.3)
    plt.xlabel("x-axis")
    plt.ylabel("y-axis")
    logger.info("labeling the x and y axis")
    plt.legend()
    logger.info("labeling plot label")
    plt.title("Popularity Of Programming Languages")

    plt.show()


def main():
    
    logger.info("Inside main method")

    data={
        'programming':['java','python','PHP','Javascript','C#','C++'],
        'popularity':[22.2,17.6,8.8,8,7.7,6.7]
        }
    
    logger.info("calling draw_bar_chart method")
    draw_bar_chart(data)


if __name__=='__main__':
    main()
```
