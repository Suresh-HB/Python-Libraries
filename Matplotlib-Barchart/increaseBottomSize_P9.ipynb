```python
import matplotlib.pyplot as plt
import pandas as pd
import myloggingfile as mlf
logger = mlf.logger_init("increaseBottomSize_P9.ipynb")


def  increasing_bottom_margin(data):
    
    """
    Description: Function to display a vertical bar chart of the popularity of programming Languages with Increase bottom margin. 
    Parameters:
       data : Function taking the data as a parameter to draw the barchart after make the data into data Frame
    Returns:
        None:
    """
     
    logger.info("inside unifor_color mathod")

    df=pd.DataFrame(data)
    plt.figure(figsize=(10, 6))

    k=plt.bar(df['languages'],df['popularity'], color='skyblue')
    logger.debug(f"ploting the bars\n{k}")
    
    plt.subplots_adjust(bottom=0.2)
    logger.info("increased bottom margin")


    plt.title('Popularity of Programming Languages')
    plt.xlabel('Programming Languages')
    plt.ylabel('Popularity (%)')
    
    plt.show()
    logger.info("printed uniform bars")

def main():

    logger.info("Inside main method")

    data = {'languages':['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'],
             'popularity':[22.2, 17.6, 8.8, 8.0, 7.7, 6.7]
           }
    
    logger.info("calling uniform_clor method")
    increasing_bottom_margin(data)
 
if __name__=='__main__':
    main()
```
