```python
import matplotlib.pyplot as plt
import pandas as pd
import myloggingfile as mlf
import numpy as np
logger = mlf.logger_init("dataFrameForBar_P11.ipynb")

def data_frame(data):

    """
    Description: Function to display a vertical bar chart for create bar plot from a DataFrame
    Parameters:
       data : Function taking the data as a parameter to draw the barchart after make the data into data Frame
    Returns:
        None:
    """

    logger.info("inside data_frame method")

    indexing=(len(data['a']))
    df = pd.DataFrame(data)
    logger.debug(f"printing the data frame \n {df}")

    position=np.arange(indexing)
    df.set_index(position, inplace=True)

    df.plot(kind='bar', figsize=(10, 6), width=0.8)

    plt.title('Bar Plot from DataFrame')
    plt.xlabel('Index (a)')
    plt.ylabel('Values')
    plt.legend(title='Columns')

    plt.show()


def main():
    
    logger.info('inside main method')

    data = {
            'a':[2, 4, 8, 5, 7, 6],
            'b':[4, 2, 3, 4, 2, 6],
            'c':[6, 4, 7, 4, 7, 8],
            'd':[8, 2, 6, 4, 8, 6],
            'e':[10,2, 4, 3, 3, 2]
           }
    
    logger.info("calling data_frame method")
    data_frame(data)

if __name__=='__main__':
    main()
```
