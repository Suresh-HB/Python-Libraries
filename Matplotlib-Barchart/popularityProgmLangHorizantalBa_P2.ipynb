```python
import matplotlib.pyplot as plt
import pandas as pd
import myloggingfile as mlf
logger = mlf.logger_init("popularityProgmLangHorizantalBa_P2.ipynb")


def draw_bar_chart(data):

    """
    Description: Function to display a horizantal bar chart of the popularity of programming Languages. 
    Parameters:
       data : Function taking the data as a parameter to draw the barchart after make the data into data Frame
    Returns:
        None:
    """

    logger.info("inside draw_bar_chart")
    df = pd.DataFrame(data)
    logger.info(f" Data Frame\n{df}")

    plt.barh(df['programming'],df['popularity'],label='popularity',color='green')
    plt.xlabel("x-axis")
    plt.ylabel("y-axis")
    logger.info("labeling the x and y axis")
    plt.legend()
    logger.info("labeling plot label")
    plt.title("Popularity Of Programming Languages")

    plt.show()


def main():
    
    logger.info("Inside main method")

    data={
        'programming':['java','python','PHP','Javascript','C#','C++'],
        'popularity':[22.2,17.6,8.8,8,7.7,6.7]
        }
    
    logger.info("calling draw_bar_chart method")
    draw_bar_chart(data)


if __name__=='__main__':
    main()
```

```python
import matplotlib.pyplot as plt
import numpy as np
 
# create data
x = ['A', 'B', 'C', 'D']
y1 = np.array([10, 20, 10, 30])
y2 = np.array([20, 25, 15, 25])
y3 = np.array([12, 15, 19, 6])
y4 = np.array([10, 29, 13, 19])
 
# plot bars in stack manner
plt.bar(x, y1, color='r')
plt.bar(x, y2, bottom=y1, color='b')
plt.bar(x, y3, bottom=y1+y2, color='y')
plt.bar(x, y4, bottom=y1+y2+y3, color='g')
plt.xlabel("Teams")
plt.ylabel("Score")
plt.legend(["Round 1", "Round 2", "Round 3", "Round 4"])
plt.title("Scores by Teams in 4 Rounds")
plt.show()
```
