```python
import matplotlib.pyplot as plt
import pandas as pd
import myloggingfile as mlf
import numpy as np
logger = mlf.logger_init("specifyingThePositionOfBar_P7.ipynb")

def position_method(data):

    """
    Description: Function to display a vertical bar chart of the popularity of programming Language 
                and Specifying the position of each bar plot. 
    Parameters:
       data : Function taking the data as a parameter to draw the barchart after make the data into data Frame
    Returns:
        None:
    """

    logger.info("inside 'position_method' method ")
    df = pd.DataFrame(data)

    positions = np.arange(len(df['languages']))

    plt.figure(figsize=(10, 6))
    bars = plt.bar(positions, df['popularity'], color='skyblue', edgecolor='blue')

    plt.xticks(positions, df['languages'])

    plt.title('Popularity of Programming Languages')
    plt.xlabel('Programming Languages')
    plt.ylabel('Popularity (%)')

    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.5, f'{yval}', ha='center', va='bottom')
    plt.show()
    

def main():
    
    logger.info("inside main method")

    data={'languages':['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'],
           'popularity':[22.2, 17.6, 8.8, 8.0, 7.7, 6.7]
        }
    
    position_method(data)

if __name__=='__main__':
    main()
```
