```python
import matplotlib.pyplot as plt
import pandas as pd
import myloggingfile as mlf
import numpy as np

logger = mlf.logger_init("plotByGroup_P10.ipynb")

def group_gender(groups,means_men,means_women):

    logger.info("inside group_gender method")

    """
    Description: Function to Python program to create bar plot of scores by group and gender. Use multiple X
                values on the same chart for men and women.
    Parameters:
       data : Function taking the data as a parameter to draw the barchart after make the data into data Frame
    Returns:
        None:
    """


    num_groups = len(groups)

    positions = np.arange(num_groups)
    bar_width = 0.4  

    plt.figure(figsize=(10, 6))

    bars_men = plt.bar(positions - bar_width / 2, means_men, width=bar_width, label='Men', color='skyblue', edgecolor='blue')
    bars_women = plt.bar(positions + bar_width / 2, means_women, width=bar_width, label='Women', color='salmon', edgecolor='red')

    logger.info("plotted the bars")
    plt.title('Scores by Group and Gender')
    plt.xlabel('Groups')
    plt.ylabel('Scores')
    plt.xticks(positions, groups) 
    logger.info("labeling the axiss and title,xticks") 

    plt.legend()
    plt.show()

def main():
    
    logger.info("inside main method")

    groups = ['Group 1', 'Group 2', 'Group 3', 'Group 4', 'Group 5']
    means_men = [22, 30, 35, 35, 26]
    means_women = [25, 32, 30, 35, 29]

    logger.info("calling group_gender method")
    group_gender(groups,means_men,means_women)

if __name__ == "__main__":
    main()
```
