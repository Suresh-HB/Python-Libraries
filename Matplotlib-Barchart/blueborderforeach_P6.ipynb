```python
import matplotlib.pyplot as plt
import pandas as pd
import myloggingfile as mlf
logger = mlf.logger_init("blueborderforeach_P6.ipynb")

def labeling_every_bar(data):
    
    """
    Description: Function to display a vertical bar chart of the popularity of programming Languages using
                and Make blue border to each bar. 
    Parameters:
       data : Function taking the data as a parameter to draw the barchart after make the data into data Frame
    Returns:
        None:
    """

    logger.info("inside labeling method")

    df = pd.DataFrame(data)
    logger.debug(f"Data frame \n{df}")

    plt.figure(figsize=(10, 6))
    bars = plt.bar(df['languages'], df['popularity'], color='yellow',edgecolor='blue')
    logger.debug(f"bars \n{bars}")

    plt.title('Popularity of Programming Languages')
    plt.xlabel('Programming Languages')
    plt.ylabel('Popularity (%)')
    logger.info("tittle and labeling the x & y axis")

    plt.show()

def main():

    logger.info("Insdie main method")

    data = {'languages':['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'],
             'popularity':[22.2, 17.6, 8.8, 8.0, 7.7, 6.7]
            }
    
    logger.info("calling labeling method")
    labeling_every_bar(data)


if __name__=='__main__':
    main()
```
