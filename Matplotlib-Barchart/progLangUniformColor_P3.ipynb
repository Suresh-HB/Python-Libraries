```python
import matplotlib.pyplot as plt
import pandas as pd
import myloggingfile as mlf
logger = mlf.logger_init("progLangUniformColor_P3.ipynb")


def uniform_clor(data):
    
    """
    Description: Function to display a vertical bar chart of the popularity of programming Languages using uniform color. 
    Parameters:
       data : Function taking the data as a parameter to draw the barchart after make the data into data Frame
    Returns:
        None:
    """
     
    logger.info("inside unifor_color mathod")

    df=pd.DataFrame(data)
    plt.figure(figsize=(10, 6))

    k=plt.bar(df['languages'],df['popularity'], color='skyblue')
    logger.debug(f"ploting the bars\n{k}")

    plt.title('Popularity of Programming Languages')
    plt.xlabel('Programming Languages')
    plt.ylabel('Popularity (%)')
    
    plt.show()
    logger.info("printed uniform bars")

def main():

    logger.info("Inside main method")

    data = {'languages':['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'],
             'popularity':[22.2, 17.6, 8.8, 8.0, 7.7, 6.7]
           }
    
    logger.info("calling uniform_clor method")
    uniform_clor(data)
 
if __name__=='__main__':
    main()
```
