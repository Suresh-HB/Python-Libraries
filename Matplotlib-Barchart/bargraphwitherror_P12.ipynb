```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import myloggingfile as mlf
logger = mlf.logger_init("bargraphwitherror_P12.ipynb")




def error_graph(mean_velocity,std_deviation,labels):

    """
    Description: Function to create bar plots with error bars on the same figure.
    Parameters:
       mean_velocity, std_deviation, labels : Function taking parameters for plotting bars
    Returns:
        None:
    """

    logger.info("inside error_graph method")
    num_bars = len(mean_velocity)
 
    positions = np.arange(num_bars)
    logger.debug(f"positions \n{positions}")
    bar_width = 0.5
    logger.debug(f"bar width \n{bar_width}")
    plt.figure(figsize=(10, 6))
 
    k=plt.bar(positions, mean_velocity, yerr=std_deviation, width=bar_width, color='skyblue', edgecolor='black', capsize=5, label='Mean Velocity')
    logger.debug(f"plotting the bars \n{k}")
    plt.title('Bar Plot with Error Bars')
    plt.xlabel('Measurements')
    plt.ylabel('Mean Velocity')
    plt.xticks(positions, labels)

    logger.debug(f"title and labeling x and y axis")
    plt.legend()
    plt.show()
    logger.debug(f"ploting the result")


def main():
  
    logger.info("inside main method")

    mean_velocity = [0.2474, 0.1235, 0.1737, 0.1824]
    std_deviation = [0.3314, 0.2278, 0.2836, 0.2645]
    labels = ['Measure 1', 'Measure 2', 'Measure 3', 'Measure 4']
    logger.info("calling error_graph method")
    error_graph(mean_velocity,std_deviation,labels)

if __name__ == "__main__":
    main()
```
