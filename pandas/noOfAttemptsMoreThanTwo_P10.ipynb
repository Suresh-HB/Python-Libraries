```python
import pandas as pd
import numpy as np
import myloggingfile as lf

logger = lf.logger_init("noOfAttemptsMoreThanTwo_P10.ipynb")

def attempts_greater_than_two(exam_data,labels):
    
    """
    Description: Function to select the rows where the number of attempts in the examination is greater than 2..
    Parameters: 
    exam_data,labels : Function taking given data and it's lables for get the specified columns and rows. 
    Returns:
        None:
    """

    logger.info(" Inside attempts_greater_than_two method")
    df = pd.DataFrame(exam_data)
    df.index=labels

    iltered_df = df[df['attempts']>2]
    print(iltered_df)


def main():

    logger.info("Inside main method")

    exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew',
    'Laura', 'Kevin', 'Jonas'],
    'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
    'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
    labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

    logger.debug(f" given data set \n{exam_data}")
    logger.debug(f" given labels \n{labels}")
    
    logger.info("calling attempts_greater_than_two method")
    attempts_greater_than_two(exam_data,labels)

if __name__ == '__main__':
    main()
```
