```python
import matplotlib.pyplot as plt
import numpy as np
import myloggingfile as mlf
logger = mlf.logger_init("diffStyleUsingArray_P11.ipynb")

def formate_style():

    """
    Description : Function to plot qu.
    Parameters : 
        data : Function that converts the data into a tabular format and then uses this
                data to plot the x-axis and y-axis.
    Returns :
        None :
    """

    logger.info("formate_style method")

    # Define the data for plotting
    x = np.linspace(0, 10, 100)
    
    y1 = np.sin(x)
    y2 = np.cos(x)
    y3 = np.tan(x)

    plt.plot(x, y1, 'r--', label='sin(x)')  # Red dashed line
    plt.plot(x, y2, 'g-.', label='cos(x)')  # Green dash-dot line
    plt.plot(x, y3, 'b:', label='tan(x)')   # Blue dotted line

    plt.title('Multiple Lines with Different Formats')
    plt.xlabel('x-axis')
    plt.ylabel('y-axis')
    
    logger.info("plotting the lines")
    logger.info("labeling the labels ")
    plt.legend()
    plt.show()

def main():
    
    logger.info("inside main method")
    formate_style()

if __name__ == '__main__':
    main()
```
