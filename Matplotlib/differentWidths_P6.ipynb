```python
import matplotlib.pyplot as plt
import pandas as pd
import myloggingfile as mlf

logger = mlf.logger_init("differentWidths_P6.ipynbs")

def draw_line(data):

    """
    Description : Function to read the data from data set and plotting the lines with different widths and colors.
    Parameters : 
        data : Function taking data and converting that data into tabular form and then reading the data
               for plotting x axis and y axis.
    Returns :
        None :
    """

    logger.info("inside draw_line method")
    
    df = pd.DataFrame(data)

    plt.plot(df['Name'],df['Age'], color='blue', linewidth = 8,  label = 'line1-width-3')
    plt.plot(df['Name'],df['Company'], color='yellow', linewidth = 3,  label = 'line1-width-3')
    plt.plot(df['Name'],df['Days'], color='green', linewidth = 3,  label = 'line1-width-3')

    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.legend()
    plt.grid()
    plt.show()


def main():

    logger.info("insdie main method")

    data = {
            'Name':['suresh','mahesh','tarun','sanju','sitaraman'],
            'Age':[23, 22, 23, 22, 18],
            'Days':[5, 8, 9, 6,3],
            'Company':['Apexon','QSG','Apexon','Apexon','info']
            }
    
    logger.info("calling draw_line method")
    draw_line(data)

if __name__ == '__main__':
    main()
```
