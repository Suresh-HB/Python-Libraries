```python
import matplotlib.pyplot as plt
import myloggingfile as mlf

logger = mlf.logger_init("drawLine_P2.ipynb")

def main():

    logger.info(" Inside main method")
    
    x = [10,30,50,70]
    y = [2,402,650,20]
    plt.plot(x, y,)

    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Line")
    plt.legend()

    plt.show()

if __name__ =='__main__':
    main()
```
