```python
import matplotlib.pyplot as plt
import pandas as pd
import myloggingfile as mlf

logger = mlf.logger_init("multipleCharts_P12.ipynb")

import matplotlib.pyplot as plt

def multiple_chart():
    
    """
    Description : Function to create multiple types of charts
    Parameters : 
    Returns :
        None :
    """
    logger.info("inside the multiple_chart")
    # Create a 2x2 grid of subplots
    fig, axs = plt.subplots(2, 2)
    
    # Now axs is a 2D array of Axes objects
    axs[0, 0].plot([1, 2, 3], [4, 5, 6])
    axs[0, 1].scatter([1, 2, 3], [4, 5, 6])
    axs[1, 0].bar([1, 2, 3], [4, 5, 6])
    axs[1, 1].hist([1, 2, 2, 3, 3, 3, 3, 3, 3, 4, 4, 5])
    
    plt.show()



def main():

    logger.info("inside main")
    multiple_chart()
if __name__=='__main__':
    main()
```
