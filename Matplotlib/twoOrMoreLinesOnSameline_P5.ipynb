```python
import matplotlib.pyplot as plt
import pandas as pd
import myloggingfile as mlf

logger = mlf.logger_init('twoOrMoreLinesOnSameline_P5.ipynb')

def data_frame(data):
    
    """
    Description : Function to read the data from data set and plotting the lines.
    Parameters : 
        data : Function taking data and converting that data into tabular form and then reading the data
               for plotting x axis and y axis.
    Returns :
        None :
    """

    logger.info("insdie data_frame methoad")
    df = pd.DataFrame(data)
    logger.debug(f"\n {df}")
    
    plt.plot(df['x'], df['line1'], label='Line 1', marker='o')
    plt.plot(df['x'], df['line2'], label='Line 2', marker='s')
    plt.plot(df['x'], df['line3'], label='Line 3', marker='^')
    logger.info("plotting the multiple lines")

    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')

    plt.legend()
    plt.show()


def main():
    
    logger.info("inside main method")
    data = { 
        'x': [1, 2, 3, 4, 5],
        'line1': [2, 3, 3, 7, 11],
        'line2': [1, 4, 4, 8, 10],
        'line3': [3, 6, 6, 12, 15]
        }
    logger.info("calling data_frame method")
    data_frame(data)

if __name__ == '__main__':
    main()

```
