```python
import matplotlib.pyplot as plt
import pandas as pd
import myloggingfile as mlf

logger = mlf.logger_init("plotQuantities_P10.ipynb")

def draw_line(x,y):

    """
    Description : Function to plot quantities which have an x and y position.
    Parameters : 
        x,y : Function that converts the data into a list format and then uses this
                data to plot the x-axis and y-axis.
    Returns :
        None :
    """

    logger.info("inside draw_line method")
    
    plt.figure(figsize=(8, 6))
    plt.scatter(x, y, color='blue', marker='o', s=100, edgecolor='black')

    logger.info("plotting the results")
    plt.title('Scatter Plot of Quantities')

    plt.xlabel('X Position')
    plt.ylabel('Y Position')
    
    logger.info("labellig the x-axis and y-axis")
    plt.grid(True)
    plt.show()


def main():

    logger.info("insdie main method")

    x=[1, 2, 3, 4, 5],
    y=[10, 15, 13, 17, 12]
    logger.info("calling draw_line method")

    draw_line(x,y)

if __name__ == '__main__':
    main()
```
