```python
import matplotlib.pyplot as plt
import pandas as pd
import myloggingfile as mlf

logger = mlf.logger_init("setLineMarks_P8.ipynb")

def draw_line(data):

    """
    Description : Function to plot two or more lines with different styles
    Parameters : 
        data : Function that converts the data into a tabular format and then uses this
                data to plot the x-axis and y-axis.
    Returns :
        None :
    """

    logger.info("inside draw_line method")
    
    df = pd.DataFrame(data)
    logger.debug(f"\n{df}")

    plt.plot(df['Score'],df['Days'], color='blue', label ='Days', marker='o')
    plt.plot(df['Score'],df['overs'], color='green', label='overs', marker='-.')
    logger.info("plotting the results")

    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    logger.info("labellig the x-axis and y-axis")

    plt.legend()
    plt.grid()
    plt.show()


def main():

    logger.info("insdie main method")

    data = {
            'Score':[6, 6, 9, 22, 18],
            'overs':[12, 20, 50, 16,30],
            'Days':[2, 6, 9, 6,3],
            }
    
    logger.info("calling draw_line method")
    draw_line(data)

if __name__ == '__main__':
    main()
```
