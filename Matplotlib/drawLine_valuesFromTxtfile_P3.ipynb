```python
import matplotlib.pyplot as plt
import myloggingfile as mlf 

logger = mlf.logger_init("drawLine_valuesFromTxtfile_P3.ipynb")


def read_data(file_path):

    """
    Description : Function to read the data from text file based on that values draws line
    Parameters : 
        file_path : Function taking text file path and reads in that data
    Returns :
        x, y : returning the list of file data  
    """

    logger.info("inside read_data method")
    
    x = []
    y = []
    with open(file_path, 'r') as file:
        logger.info("readding the data from txt file")
        for line in file:
            values = line.split()
            if len(values) == 2:
                x.append(int(values[0]))
                y.append(int(values[1]))
    return x, y


def main():
    
    logger.info("Inside main method")

    file_path = 'test.txt'

    logger.info(" calling read data")
    x, y = read_data(file_path)
    plt.plot(x, y, marker='*', linestyle='-', color='b')


    plt.xlabel('X Axis')
    plt.ylabel('Y Axis')
    plt.title('Line Plot from Test Data')

    logger.info("labled with x axis and y axis")
    logger.info("title is given")
    
    plt.grid(True)
    plt.show()

    logger.info("plot results")


if __name__ == '__main__':
    main()
```
